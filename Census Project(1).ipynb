Correct the column names and translate

```python
newdf.columns=['Level', 'Geographic Key', 'Description',
          'Sum of both sexes', '0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34',
          '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', 
          '75-79', '80-84', '85+',
          
          'Sum of Males', 'M 0-4', 'M 5-9', 'M 10-14', 'M 15-19', 'M 20-24', 'M 25-29', 'M 30-34', 'M 35-39',
            'M 40-44', 'M 45-49', 'M 50-54', 'M 55-59', 'M 60-64', 'M 65-69', 'M 70-74', 'M 75-79', 'M 80-84', 'M 85+',
         
          'F sum', 'F 0-4', 'F 5-9', 'F 10-14', 'F 15-19', 'F 20-24', 'F 25-29', 'F 30-34', 'F 35-39', 'F 40-44',
               'F 45-49', 'F 50-54', 'F 55-59', 'F 60-64', 'F 65-69', 'F 70-74', 'F 75-79', 'F 80-84', 'F 85+']
```

```python
# Translate what can be transalted, the rest is better left in Greek as it is mostly NAMES. Assign everything to the temp list.
temp_list=[]

for desc in newdf['Description'].values:
    desc=desc.split()
    if desc[0]== 'ΔΗΜΟΣ':
        desc[0]=desc[0].replace('ΔΗΜΟΣ','Municipality')
    elif desc[0]== 'ΔΗΜΟΤΙΚΗ' and desc[1]=='ΕΝΟΤΗΤΑ':
        desc[0]=desc[0].replace('ΔΗΜΟΤΙΚΗ','Municipal')
        desc[1]=desc[1].replace('ΕΝΟΤΗΤΑ','Unit')
    elif desc[0]== 'ΠΕΡΙΦΕΡΕΙΑΚΗ':
        desc[0]=desc[0].replace('ΠΕΡΙΦΕΡΕΙΑΚΗ','Regional')
        desc[1]=desc[1].replace('ΕΝΟΤΗΤΑ','Unit')
    elif desc[0]== 'ΠΕΡΙΦΕΡΕΙΑ':
        desc[0]=desc[0].replace('ΠΕΡΙΦΕΡΕΙΑ','Region')
    elif desc[0]== 'ΑΠΟΚΕΝΤΡΩΜΕΝΗ':
        desc[0]=desc[0].replace('ΑΠΟΚΕΝΤΡΩΜΕΝΗ','Decentralized')
        desc[1]=desc[1].replace('ΔΙΟΙΚΗΣΗ','Administration')
    elif desc[0]=='ΔΗΜΟΤΙΚΕΣ':
        desc[0]=desc[0].replace('ΔΗΜΟΤΙΚΕΣ','Municipal')
        desc[1]=desc[1].replace('ΕΝΟΤΗΤΕΣ','Units')
    elif '(ΑΥΤΟΔΙΟΙΚΗΤΟ)' in desc:
        desc[-1]=desc[-1].replace('(ΑΥΤΟΔΙΟΙΚΗΤΟ)', 'Self-Governing')
    if 'ΚΑΙ' in desc:
        temp_index=desc.index('ΚΑΙ')
        desc[temp_index]=desc[temp_index].replace('ΚΑΙ', 'And')
    temp_list.append(' '.join(desc))
    
newdf['Geographic Key'][0]= 0
newdf['Description'].values[0]= 'Country Total'
#replace the descriptions with the translated 
newdf['Description']= temp_list
```

Visualise the proportions of the population that are Males or Females accordingly

```python

import numpy as np

# Sample data
categories = ['Patras', 'Thessaloniki', 'Athens']
males = [males_patras, males_thesalloniki, males_athens]
females = [females_patras, females_thesalloniki, females_athens]
# Create an array for the x-axis positions
x = np.arange(len(categories))

# Plotting the stacked bar chart for males and females in each population
fig, ax = plt.subplots()

# Plotting females
ax.bar(x, females, color='red', label='Females')

# Plotting males on top of females
ax.bar(x, males, bottom=females, color='blue', label='Males')
percentage_of_males=[] #initialize a list to count the percentages
for i in range(len(categories)):
    total = males[i] + females[i]
    percentage_of_males.append(round(males[i]/total*100,2))
    ax.text(x[i], total/2*1.2, str(percentage_of_males[i]), ha='center', va='center', color='white', fontweight='bold')

labels=['Female','Male']
handles= ax.get_legend_handles_labels()[0]

plt.xlabel('Municipalities')
plt.ylabel('Total Population')
plt.title('Population Distribution (Males and Females) Across Different Municipalities')
plt.xticks(x, categories)
plt.legend(reversed(handles), reversed(labels), title="Legend", loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Find the total population and assign it to a variable.
total_population=newdf[Municipalities]['Sum of both sexes'].values.sum()
```
